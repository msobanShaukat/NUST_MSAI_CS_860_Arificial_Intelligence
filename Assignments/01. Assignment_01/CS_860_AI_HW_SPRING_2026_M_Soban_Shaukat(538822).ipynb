### üìã Problem Statement

Write a program to construct the following **star pattern** using **nested for loops**.

Pattern (n=5)
‚úÖ **Requirement:** Print stars in ascending order, then descending order   

---

````python
# ============================================
# Task 12: Construct a Star Pattern
# ============================================

n = 5

# Step 1: Print ascending pattern
for i in range(1, n + 1):
    for j in range(i):
        print("*", end=" ")
    print()  # move to next line

# Step 2: Print descending pattern
for i in range(n - 1, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
````

### üìã Problem Statement

Write a program to check the **validity of passwords** input by users.

‚úÖ **Validation Rules:**
- At least 1 letter between [a-z]
- At least 1 letter between [A-Z]
- At least 1 number between [0-9]
- At least 1 character from [$#@]
- Minimum length: 6 characters
- Maximum length: 16 characters

---

````python
# ============================================
# Task 19: Check Password Validity
# ============================================

# Step 1: Input password from user
password = input("Enter your password: ")

# Step 2: Initialize flags
has_lower = False
has_upper = False
has_digit = False
has_special = False

# Step 3: Check each character using a loop
for ch in password:
    if ch.islower():
        has_lower = True
    elif ch.isupper():
        has_upper = True
    elif ch.isdigit():
        has_digit = True
    elif ch in "$#@":
        has_special = True

# Step 4: Check all conditions
if len(password) < 6 or len(password) > 16:
    print("Invalid password: length must be between 6 and 16 characters.")
elif not has_lower:
    print("Invalid password: must contain at least one lowercase letter [a-z].")
elif not has_upper:
    print("Invalid password: must contain at least one uppercase letter [A-Z].")
elif not has_digit:
    print("Invalid password: must contain at least one digit [0-9].")
elif not has_special:
    print("Invalid password: must contain at least one special character [$#@].")
else:
    print("Valid password!")
````

### üìã Problem Statement  
Write a Python function that checks whether a given string is a **palindrome**.  
A palindrome reads the same backward as forward (e.g., *madam*, *nurses run*).

### ‚úÖ Requirement  
- Ignore spaces and case  
- Compare original and reversed string  

---


````python
# ============================================
# Task 25: Check Whether a String is a Palindrome
# ============================================

# Step 1: Define a function to check for palindrome
def is_palindrome(s):
    # Remove spaces and convert to lowercase
    s = s.replace(" ", "").lower()

    # Reverse the string manually using a loop
    reversed_str = ""
    for char in s:
        reversed_str = char + reversed_str

    # Check if original and reversed are the same
    if s == reversed_str:
        print("It is a palindrome.")
    else:
        print("It is not a palindrome.")

# Step 2: Example usage
text = input("Enter a word or phrase: ")
is_palindrome(text)
````


## üéØ TASK 26: Print Pascal‚Äôs Triangle

### üìã Problem Statement  
Write a Python function that prints the **first n rows** of Pascal‚Äôs triangle.  
Each number is the sum of the two numbers directly above it.

### ‚úÖ Requirement  
- Use a function  
- Use loops to generate rows  

---

````python
# ============================================
# Task 26: Print Pascal‚Äôs Triangle
# ============================================

# Step 1: Define function to print Pascal's triangle
def pascal_triangle(n):
    for i in range(n):
        num = 1
        for j in range(i + 1):
            print(num, end=" ")
            num = num * (i - j) // (j + 1)
        print()

# Step 2: Input number of rows
rows = int(input("Enter number of rows: "))

# Step 3: Call the function
pascal_triangle(rows)
````


## üéØ TASK 27: Check Whether a String is a Pangram

### üìã Problem Statement  
Write a Python function to check whether a given sentence is a **pangram**.  
A pangram contains **every letter** of the alphabet at least once.

**Example:**  
‚ÄúThe quick brown fox jumps over the lazy dog‚Äù

### ‚úÖ Requirement  
- Convert input to lowercase  
- Check each letter a‚Äìz  

---


````python
# ============================================
# Task 27: Check Whether a String is a Pangram
# ============================================

# Step 1: Define a function to check pangram
def is_pangram(sentence):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    sentence = sentence.lower()

    for letter in alphabet:
        if letter not in sentence:
            return False
    return True

# Step 2: Example usage
text = input("Enter a sentence: ")

# Step 3: Display result
if is_pangram(text):
    print("The sentence is a pangram.")
else:
    print("The sentence is not a pangram.")
````

````python
# ============================================
# Task 28: Invoke a Function After a Delay
# ============================================

# Step 1: Function to calculate square root manually (Newton-Raphson method)
def manual_sqrt(number):
    guess = number / 2
    for i in range(10):
        guess = 0.5 * (guess + number / guess)
    return guess

# Step 2: Function to simulate delay (without using time.sleep)
def delay(milliseconds):
    count = 0
    for i in range(milliseconds * 1000):
        count += 1

# Step 3: Function to invoke after delay
def delayed_square_root(number, delay_ms):
    print(f"Square root after {delay_ms} milliseconds:")
    delay(delay_ms)
    print(manual_sqrt(number))

# Step 4: Example usage
delayed_square_root(16, 1)
delayed_square_root(100, 1)
delayed_square_root(25100, 1)
````


## üéØ TASK 29: Find Prime Factors of a Number

### üìã Problem Statement  
Write a Python program to find the **prime factors** of a number.  
Example: prime factors of 56 ‚Üí 2, 2, 2, 7

### ‚úÖ Requirement  
- Use a loop  
- Divide the number by smallest prime factors  

---

````python
# ============================================
# Task 29: Find Prime Factors of a Number
# ============================================

# Step 1: Define a function to find prime factors
def prime_factors(n):
    i = 2
    factors = []
    while i <= n:
        if n % i == 0:
            factors.append(i)
            n = n // i
        else:
            i += 1
    return factors

# Step 2: Input number from user
num = int(input("Enter a number: "))

# Step 3: Display prime factors
print("Prime factors of", num, "are:", *prime_factors(num))
````


## üéØ TASK 30: Convert Decimal to Binary

### üìã Problem Statement  
Write a Python function that converts a **decimal number** into a **binary string**.

### ‚úÖ Requirement  
- Use a loop and division by 2  
- Build the binary string  

---

````python
# ============================================
# Task 30: Convert Decimal to Binary
# ============================================

# Step 1: Define function to convert decimal to binary
def decimal_to_binary(n):
    binary = ""
    if n == 0:
        return "0"
    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n = n // 2
    return binary

# Step 2: Input number from user
num = int(input("Enter a decimal number: "))

# Step 3: Display binary result
print("Binary number:", decimal_to_binary(num))
````


## üéØ BONUS: Armstrong Numbers Using cubesum(), PrintArmstrong(), and isArmstrong()

### üìã Problem Statement  
Write a function **cubesum()** that returns the sum of cubes of digits of a number.  
Use it to build:

- **isArmstrong()** ‚Üí checks if a number is Armstrong  
- **PrintArmstrong()** ‚Üí prints all Armstrong numbers in a range  

### ‚úÖ Requirement  
- Use digit extraction  
- Use cubesum() inside other functions  

---

````python
# ============================================
# Bonus: Armstrong Numbers
# ============================================

# Step 1: Function to calculate sum of cubes of digits
def cubesum(n):
    total = 0
    while n > 0:
        digit = n % 10
        total += digit ** 3
        n = n // 10
    return total

# Step 2: Function to check whether a number is Armstrong
def isArmstrong(num):
    return num == cubesum(num)

# Step 3: Function to print Armstrong numbers in a range
def PrintArmstrong(start, end):
    print("Armstrong numbers between", start, "and", end, "are:")
    for i in range(start, end + 1):
        if isArmstrong(i):
            print(i)

# Step 4: Example usage
num = int(input("Enter a number: "))
if isArmstrong(num):
    print(num, "is an Armstrong number.")
else:
    print(num, "is not an Armstrong number.")

# Print Armstrong numbers between 1 and 500
PrintArmstrong(1, 500)
````
